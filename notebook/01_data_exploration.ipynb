```python
import json
```

# Data load

```python
with open("../data/chat.txt", "r", encoding="utf-8") as f:
    data = f.readlines()
```

```python
data[0]
```

Agrupación de datos en chats logicos, o batches

```python
clean_data = []

last_name_user = ""
chunk_data = []

for line in data[1:]:
    line = line.strip()
    line = line.split("-")[1:]
    line = "-".join(line)

    name_user = line.split(":")[0].strip()
    text = ":".join(line.split(":")[1:]).strip()

    # print(text)

    if text == "<Multimedia omitido>":
        continue

    if last_name_user != name_user:
        if last_name_user == "":
            last_name_user = name_user
            continue

        clean_data.append(last_name_user + ":"  + "\n".join(chunk_data))

        chunk_data = []
        last_name_user = name_user
        chunk_data.append(text)
    else:
        chunk_data.append(text)
```

```python
# Formación de QA

qa_data = []
data_init = {}

for data in clean_data:
    name_user = data.split(":")[0].strip()
    text = ":".join(data.split(":")[1:]).strip()

    if name_user == "Miguel":
        data_init["answer"] = text
    elif name_user == "Ing Shelryn La Junior":
        data_init["question"] = text


    if len(data_init) == 2:
        qa_data.append(data_init)
        data_init = {}
```

```python
with open("../data/qa_data.json", "w", encoding="utf-8") as f:
    json.dump(qa_data, f, ensure_ascii=False, indent=4)
```
